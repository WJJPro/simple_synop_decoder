```python
def decMsgd1_1(msg: str):
    raw = msg.split(' ')
    res = []
    # 第一段电文
    # II iii
    # II为区号 iii为站号
    code1 = raw[0]
    II = code1[0:2]
    iii = code1[2:5]
    res += ['区号:{} 站号:{}'.format(II,iii)]
    
    # 第二段电文
    # i_R i_X h v v
    # i_R i_X为指示码
    # h 为最低的云的高度
    # vv为有效能见度
    code2 = raw[1]
    i_Ri_X = code2[0:2]
    h = code2[2]
    vv = int(code2[3:5])
    if vv <= 50:
        vv = vv * 0.1
    elif vv <= 55:
        vv = None
    elif vv <= 80:
        vv = vv - 50
    elif vv <= 89:
        vv = 30 + (vv - 80) * 5
    else:
        vv = vv_map[vv-90]
    res += ['指示码:{} 最低的云的高度={}m 有效能见度={}km'.format(i_Ri_X,h,vv)]
    
    # 第3段电文
    # N dd ff
    # N为总云量
    # dd为风向
    # ff为风速
    code3 = raw[2]
    N = int(code3[0])
    dd = int(code3[1:3]) # 单位10度
    ff = code3[3:5]
    
    res += ['总云量:{} 风向:{} 风速:{:.1f}m/s'.format(N_map[N],dd*10,float(ff))]
    return res
```

```python
def decMsgd1_2(msg: str):
    tmp = ''
    if msg[0] == '1':
        # 气温
        # 1 S_n T T T
        tmp += "气温="
        if msg[1] == '0':
            tmp += '+'
        elif msg[1] == '1':
            tmp += '-'   
        T = float(msg[2:]) #单位0.1摄氏度
        tmp += "{:.1f}摄氏度".format(T/10.0) # 温度数值,单位1摄氏度
    elif msg[0] == '2':
        # 露点温度
        # 2 S_n T_d T_d T_d
        tmp += "露点温度=" 
        if msg[1] == '0':
            tmp += '+'
        elif msg[1] == '1':
            tmp += '-'
        T = float(msg[2:]) #单位0.1摄氏度
        tmp += "{:.1f}摄氏度".format(T/10.0) # 温度数值,单位1摄氏度
    elif msg[0] == '3':
        # 本站气压
        # 3 P_0 P_0 P_0 P_0
        tmp += "本站气压="
        pressure = float(msg[2:])/10
        if pressure <= 50:
            pressure += 1000
        else:
            pressure += 900
        tmp += "{:.1f}hpa".format(pressure)
    elif msg[0] == '4':
        # 海平面气压
        # 4 P P P P
        tmp += "海平面气压="
        pressure = float(msg[2:])/10
        if pressure <= 50:
            pressure += 1000
        else:
            pressure += 900
        tmp += "{:.1f}hpa".format(pressure)
    elif msg[0] == '5':
        # 3小时变压
        # 3 a p p p
        tmp += "3小时变压="
        a = ''
        if msg[1] == '2':
            a = '+'
        elif msg[1] == '4':
            a = '不变'
        elif msg[1] == '7':
            a = '-'
        tmp += a
        tmp += "{:.1f}hpa".format(float(msg[2:4]))
    elif msg[0] == '6':
        # 过去6小时降水量
        # 6 R R R 1
        tmp += "过去6小时降水量="
        num = int(msg[1:4])
        if num < 990 and num > 0:
            tmp += msg[1:4] + 'mm'
        elif num == 0:
            tmp += "无降水"
        elif num == 990:
            tmp +="微量"
        elif num > 990:
            tmp += '0.' + msg[4]  + 'mm'
    elif msg[0] == '7':
        # 天气现象
        # 7 w w W_1 W_2
        ww = int(msg[1:3])
        W_1 = int(msg[3])
        W_2 = int(msg[4])
        tmp += "现在天气现象:{} ".format(ww_map[ww])
        tmp += "过去6小时天气现象:{} ".format(W_map[W_1])
        tmp += "过去3小时天气现象:{}".format(W_map[W_2])
        
    elif msg[0] == '8':
        # 云描述
        # 8 N_h C_L C_M C_H
        # N_h 有C_L低云时，为C_L云的总量，否则为C_M云的总量
        # 
        tmp += "云描述:"
        
        C_L_map = ["无低云", 
                   "淡积云",
                  "浓积云",
                  "秃积雨云",
                  "积云性层积云",
                  "普通层积云",
                  "层云或碎层云",
                   "碎雨云",
                   "不同高度的积云和层积云",
                    "鬃积雨云或砧状积雨云"]
        C_M_map = ["无中云","透光高层云或雨层云","蔽光高层云","透光高积云",
                   "荚状高积云","系统发展的辐辏状高积云","积云性高积云",
                  "复高积云或蔽光高积云","堡状或絮状高积云","混乱天空的高积云"]
        C_H_map = ["无高云","毛卷云","密卷云","伪卷云","钩卷云","卷层云",
                  "卷层云","卷层云","卷层云","卷积云"]
        N_h, C_L, C_M, C_H = 0, 0, 0, 0
        if msg[1].isdigit():
            N_h = int(msg[1])
        if msg[2].isdigit():
            C_L = int(msg[2])
        if msg[3].isdigit():
            C_M = int(msg[3])
        if msg[4].isdigit():
            C_H = int(msg[4])
        
        nh = N_map[N_h]
        cl = C_L_map[C_L]
        cm = C_M_map[C_M]
        ch = C_H_map[C_H]
        
        tmp += "总云量={} ".format(nh)
        tmp += "低云:{} ".format(cl)
        tmp += "中云:{} ".format(cm)
        tmp += "高云:{} ".format(ch)
    return tmp
```

```python
def decMsgd2(msg:str):
    tmp = ''
    if msg[0] == '0':
        # 0 P_24 P_24 T_24 T_24
        # 24小时变压 24小时变温 负值绝对值加50
        P = int(msg[1:3])
        T = int(msg[3:5])
        if P >= 50:
            P -= 50
        if T >= 50:
            T -= 50
        tmp += '24小时变压={}摄氏度 '.format(P)
        tmp += '24小时变温={}摄氏度'.format(T)
    if msg[0] == '1':
        # 过去24小时最高气温
        # 1 S_n T_x T_x T_x
        tmp += "过去24小时最高气温="
        if msg[1] == '0':
            tmp += '+'
        elif msg[1] == '1':
            tmp += '-'   
        T = float(msg[2:]) #单位0.1摄氏度
        tmp += "{:.1f}摄氏度".format(T/10.0) # 温度数值,单位1摄氏度
    elif msg[0] == '2':
        # 过去24小时最低气温
        # 2 S_n T_d T_d T_d
        tmp += "过去24小时最低气温=" 
        if msg[1] == '0':
            tmp += '+'
        elif msg[1] == '1':
            tmp += '-'
        T = float(msg[2:]) #单位0.1摄氏度
        tmp += "{:.1f}摄氏度".format(T/10.0) # 温度数值,单位1摄氏度        
    elif msg[0] == '3':
        # 过去24小时最低气温
        # 3 S_n T_d T_d T_d
        tmp += "地面最低温度=" 
        if msg[1] == '0':
            tmp += '+'
        elif msg[1] == '1':
            tmp += '-'
        T = float(msg[2:]) #单位0.1摄氏度
        tmp += "{:.1f}摄氏度".format(T/10.0) # 温度数值,单位1摄氏度
    elif msg[0] == '7':
        # 过去24小时总降水量
        # 7 R_24 R_24 R_24 R_24
        num = int(msg[1:4])
        if num < 990 and num > 0:
            tmp += msg[1:4] + 'mm'
        elif num == 0:
            tmp += "无降水"
        elif num == 990:
            tmp +="微量"
        elif num > 990:
            tmp += '0.' + msg[4]  + 'mm'
    elif msg[0:3] == '911':
        # 瞬时大风风速
        # 911 fxfx
        fx = int(msg[3:5])
        tmp += '瞬时大风 极大瞬间风速={}m/s'.format(fx)
    elif msg[0:3] == '915':
        # 瞬时大风风向
        # 915 dd
        dd = int(msg[3:5]) #单位10度
        tmp += '瞬时大风 风向={}'.format(dd*10) # 单位1度
    elif msg[0:3] == '919':
        # 龙卷或尘卷风
        # 919 M_w D_a
        M_w = int(msg[3])
        D_a = int(msg[4])
        tmp += "龙卷 {} 方位{}".format(M_w_map(M_w), D_a_map(D_a))
    elif msg[0:3] == '925':
        # 积雪
        # 925 SS
        # SS积雪深度 单位厘米
        SS = int(msg[3:5])
        tmp += "积雪 深度={}".format(SS)
    elif msg[0:3] == '934':
        # 雨凇
        # 934 RR
        # RR为电线积冰直径
        # RR≤55,毫米
        # RR>55,积冰直径=(RR-50)*10
        RR = int(msg[3:5])
        rr = None
        if RR <= 55:
            rr = RR
        elif RR > 55:
            rr = (RR - 50) * 10
        tmp += "雨凇 电线积冰直径={}mm".format(rr)
    elif msg[0:3] == '939':
        # 冰雹
        # 939 nn
        # nn为最大冰雹的最大直径,单位毫米
        nn = int(msg[3:5])
        tmp += "冰雹 最大直径={}mm".format(nn)
    return tmp
        
```

```python
def decMsg(raw):
    #在?开始解码了
    if raw.find('AAXX') != -1:
        d0 = ' '.join(raw.split(' ')[:2])
        raw1 = raw[raw.find(d0)+len(d0):].strip(' ')
    else:
        d0=None
        raw1 = raw
    
    # raw1 = raw[raw.find(d0)+len(d0):].strip(' ')
    
#     d0 = ' '.join(raw.split(' ')[:2])
    d1 = raw1.split('333//')[0].strip(' ')# 1段
    d1_1 = ' '.join(d1.split(' ')[:3])
    d1_2 = ' '.join(d1.split(' ')[3:])
    d3 = raw1.split('333//')[1].strip(' ')# 2段

    if d0 != None:
        for msg, decMsg in zip(d0.split(' '), decMsgd0(d0)):
            print("电码:{},解码:{}".format(msg, decMsg))
    for msg,decMsg in zip(d1_1.split(' '),decMsgd1_1(d1_1)):
        print("电码:{},解码:{}".format(msg, decMsg))    
    for msg in d1_2.split(' '):
        print("电码:{},解码:{}".format(msg, decMsgd1_2(msg)))
    for msg in d3.split(' '):
        print("电码:{},解码:{}".format(msg, decMsgd2(msg)))
```
